```python
import pandas as pd
dados = pd.read_csv('Users.csv', sep=',')
dados.info()
```

```python
del dados['Friend_Count']
del dados['Groups_Count']
del dados['Favorite_Authors']
del dados['Detected Gender']
del dados['About-Character-Length']
```

```python
dados_br  =  dados.loc[dados['Detected Country'] == 'Brazil']
dados_br
```

```python
dados_site = dados_br.groupby(['Review_Count']).size().to_frame(name = 'count').reset_index().sort_values(by=['count'],ascending=False)
```

```python
dados_site
```

```python
dados_site.to_csv("dados.csv", sep=";")
```
